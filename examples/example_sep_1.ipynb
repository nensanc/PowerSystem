# Este Script permite definir una forma de separar los sistemas de la IEEE en varias áreas, solo utilizando los nodos de cada área. 

```python
# Cargamos las librerías
import pandapower as pp, pandapower.networks as pn
```

```python
# Configuración de los modelos
net = pn.case39()
net
```

```python
# Ajustamos el área A 
net_A = pn.case39()
bus_A = [1,2,30,37,25,3,18,26,17,27,28,29,38]
```

```python
# Seleccionamos las barras del área A
net_A.bus.loc[net_A.bus["name"].isin(bus_A)]
```

```python
# Seleccionamos las líneas del área A
bus_A_aux = [i-1 for i in bus_A]
net_A.line.loc[net_A.line["from_bus"].isin(bus_A_aux) & net_A.line["to_bus"].isin(bus_A_aux)]
```

```python
# Seleccionamos las cargas del área A 
bus_A_aux = [i-1 for i in bus_A]
net_A.load.loc[net_A.load["bus"].isin(bus_A_aux)]
```

```python
# Seleccionamos los generadores del área A 
bus_A_aux = [i-1 for i in bus_A]
net_A.gen.loc[net_A.gen["bus"].isin(bus_A_aux)]
```

```python
# Seleccionamos external grid
net_A.ext_grid.loc[net_A.ext_grid["bus"].isin(bus_A)]
```

```python
# Seleccionamos los transformadores del área A
bus_A_aux = [i-1 for i in bus_A]
net_A.trafo.loc[net_A.trafo["hv_bus"].isin(bus_A_aux) & net_A.trafo["lv_bus"].isin(bus_A_aux)]
```

```python
# filtramos por Coordenadas
bus_A_aux = [i-1 for i in bus_A]
net_A.bus_geodata.filter(items=bus_A_aux, axis=0)
```

```python
# filtramos por costos de generadores
bus_A_aux = [i-1 for i in bus_A]
net_A.poly_cost.filter(items=bus_A_aux, axis=0)
```
